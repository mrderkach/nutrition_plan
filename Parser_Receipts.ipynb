```python
fin1 = open('DB/db_dishtype.csv', 'r')
dish_dict = {}
for line in fin1:
    value, key = line.strip('\n').split(',')
    dish_dict[key] = value
#print(dish_dict)
```

```python
# Ð¤ÑƒÐ½ÐºÑ†Ð¸Ñ Ð²Ð¾Ð·Ð²Ñ€Ð°Ñ‰Ð°ÐµÑ‚ url ÐºÐ°Ñ€Ñ‚Ð¸Ð½ÐºÐ¸ Ð¸ Ñ€ÐµÑ†ÐµÐ¿Ñ‚ Ð¿Ð¾ ÑÑÑ‹Ð»ÐºÐµ 
def handle(page, name):

    imgurls = page.find('picture', {'class': 'photo-wrap'})
    if imgurls is None:
        imgurl = "None"
    else:
        imgurl = imgurls.find('source').get('srcset')


    try:
        summ_p = page.find('dl', {'class': 'summary-data'})
        summ = summ_p.find("dt").text + " " + summ_p.find("dd").text
    except:
        summ = ""
        
    try:
        ingr = ["Ingredients:"]
        for i in page.find('ul', {'class': 'ingredients'}).find_all("li", recursive=False):
            ingr += [i.text.strip()]
    except:
        ingr = [""]

        
    try:
        inst = ["Instructions:"]
        for i in page.find('div', {'class': 'instructions'}).find("ol", recursive=False).find_all("li", recursive=False):
            inst += ["\n".join([j.text.strip() for j in i.find("ol").find_all("li")])]

    except:
        inst = [""]

    return imgurl, "\n".join([summ, "\n".join(ingr), "\n".join(inst)])
```
